```python
!pip install ultralytics -q

import os
import zipfile
import urllib.request
import shutil
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# 1. Veri setini indir
dataset_url = "https://github.com/ultralytics/assets/releases/download/v0.0.0/brain-tumor.zip"
zip_path = "brain-tumor.zip"
if not os.path.exists(zip_path):
    print("⬇️ Veri seti indiriliyor...")
    urllib.request.urlretrieve(dataset_url, zip_path)

# 2. Geçici bir klasöre çıkar
print("📂 Dosyalar geçici klasöre çıkartılıyor...")
temp_extract = "/content/temp_brain"
if os.path.exists(temp_extract):
    shutil.rmtree(temp_extract)
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(temp_extract)

# 3. Kusursuz ve standart bir YOLO yapısı oluştur
print("🧹 Veri seti standart YOLO formatına dönüştürülüyor...")
base_dataset_dir = "/content/brain_tumor_dataset"
if os.path.exists(base_dataset_dir):
    shutil.rmtree(base_dataset_dir)

# Gerekli alt klasörleri oluştur
os.makedirs(f"{base_dataset_dir}/train/images", exist_ok=True)
os.makedirs(f"{base_dataset_dir}/train/labels", exist_ok=True)
os.makedirs(f"{base_dataset_dir}/val/images", exist_ok=True)
os.makedirs(f"{base_dataset_dir}/val/labels", exist_ok=True)

# Geçici klasördeki dosyaları tarayıp doğru yerlerine taşıyalım
for root, dirs, files in os.walk(temp_extract):
    for file in files:
        file_path = os.path.join(root, file)

        # Eğitim (train) verileri için yönlendirme
        if "train" in root.lower():
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                shutil.copy(file_path, f"{base_dataset_dir}/train/images/{file}")
            elif file.lower().endswith('.txt') and file != "classes.txt":
                shutil.copy(file_path, f"{base_dataset_dir}/train/labels/{file}")

        # Doğrulama (valid veya val) verileri için yönlendirme
        elif "valid" in root.lower() or "val" in root.lower():
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                shutil.copy(file_path, f"{base_dataset_dir}/val/images/{file}")
            elif file.lower().endswith('.txt') and file != "classes.txt":
                shutil.copy(file_path, f"{base_dataset_dir}/val/labels/{file}")

# 4. Standart YAML dosyasını oluştur
yaml_content = f"""
path: {base_dataset_dir}
train: train/images
val: val/images

names:
  0: negatif
  1: pozitif
"""

yaml_file_path = "brain-tumor.yaml"
with open(yaml_file_path, "w") as f:
    f.write(yaml_content)

print("📝 Kusursuz YAML haritası oluşturuldu. Eğitim başlatılıyor...")

# 5. Model Eğitimi
model = YOLO("yolov8n.pt")
results = model.train(
    data=yaml_file_path,
    epochs=50,
    imgsz=640,
    device=0,
    plots=True
)

# 6. Model Performans Ölçümü (Metrikler)
print("\n📊 Model performansı ölçülüyor...")
metrics = model.val()

precision = metrics.results_dict['metrics/precision(B)']
recall = metrics.results_dict['metrics/recall(B)']
map50 = metrics.results_dict['metrics/mAP50(B)']
map50_95 = metrics.results_dict['metrics/mAP50-95(B)']

print("\n" + "="*50)
print("🎯 ÖDEVDE İSTENİLEN DEĞERLENDİRME METRİKLERİ SONUÇLARI")
print("="*50)
print(f"🔹 Kesinlik (Precision)         : {precision:.4f}")
print(f"🔹 Duyarlılık (Recall)           : {recall:.4f}")
print(f"🔹 mAP@0.5 (Genel Doğruluk)      : {map50:.4f}")
print(f"🔹 mAP@0.5:0.95                  : {map50_95:.4f}")
print("="*50)

# 7. Çıktıların Görselleştirilmesi
train_dir = results.save_dir

confusion_matrix_path = os.path.join(train_dir, "confusion_matrix.png")
if os.path.exists(confusion_matrix_path):
    img_cm = cv2.imread(confusion_matrix_path)
    plt.figure(figsize=(7, 7))
    plt.imshow(cv2.cvtColor(img_cm, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title("Karışıklık Matrisi (Confusion Matrix)")
    plt.show()

results_png_path = os.path.join(train_dir, "results.png")
if os.path.exists(results_png_path):
    img_res = cv2.imread(results_png_path)
    plt.figure(figsize=(12, 8))
    plt.imshow(cv2.cvtColor(img_res, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title("Eğitim Süreci Kayıp ve Metrik Grafikleri")
    plt.show()

# 8. Test Görseli Üzerinde Tahmin
valid_images_dir = f"{base_dataset_dir}/val/images"
test_image_name = os.listdir(valid_images_dir)[0]
test_image_path = os.path.join(valid_images_dir, test_image_name)

best_model = YOLO(os.path.join(train_dir, "weights/best.pt"))
best_model.predict(source=test_image_path, save=True, conf=0.25)

predicted_image_path = os.path.join("runs/detect/predict", test_image_name)
if not os.path.exists(predicted_image_path):
    predict_dirs = [d for d in os.listdir("runs/detect") if d.startswith("predict")]
    predict_dirs.sort()
    predicted_image_path = os.path.join("runs/detect", predict_dirs[-1], test_image_name)

if os.path.exists(predicted_image_path):
    img_pred = cv2.imread(predicted_image_path)
    plt.figure(figsize=(6, 6))
    plt.imshow(cv2.cvtColor(img_pred, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(f"Örnek Tahmin Çıktısı ({test_image_name})")
    plt.show()

print("\n🚀 Tüm işlemler başarıyla tamamlandı!")
```
